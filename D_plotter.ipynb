```python
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.interpolate import interp1d
from matplotlib.ticker import MaxNLocator
```

## RAW

```python
labels = [
    "|v|",
    r'$\text{d}_{5}$',
    "OTO",
    "PC1 SOAP",
    r'$\text{NN}_{10}$',
    "LENS",
]
```

```python
fontsize = 18
lsize = 16
# Carica la lista delle finestre
win = np.load("results/D_arrays/windows_list.npy")
# Imposta la directory
directory = "results/ST_arrays"

# Crea una nuova figura
plt.figure()

a = 0
# Itera attraverso tutti i file nella directory
for filename in os.listdir(directory):
    if(filename.startswith("sp")):
        continue
    print(f"{a}{filename}")
    # Controlla se il file  un file .npy
    if filename.endswith(".npy") and filename != "windows_list.npy":
        # Carica i dati
        D = np.load(f"{directory}/{filename}")
        interp_func = interp1d(win, D, kind='linear')
        win_new = np.linspace(min(win), max(win), 500)
        D_interp = interp_func(win_new)
        
        # Traccia i dati
        plt.scatter(win, D, label=labels[a],s=10)
        plt.plot(win_new, D_interp)
        a += 1

# Aggiungi etichette agli assi
plt.xlabel('Time Resolution [ns]', fontweight='bold',fontsize=fontsize)
plt.xscale("log")
plt.ylabel('n. states', fontweight='bold',fontsize=fontsize)
plt.tick_params(axis="x",labelsize=lsize)
plt.tick_params(axis="y",labelsize=lsize)

# Aggiungi un titolo al grafico
plt.title('Sp. avg. number of states',fontsize=fontsize)
# Aggiungi una legenda
plt.legend(loc='lower left')

# Mostra il grafico
plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from matplotlib.ticker import MultipleLocator

# Imposta le dimensioni dei caratteri
fontsize = 18
lsize = 16

# Carica la lista delle finestre
win = np.load("results/D_arrays/windows_list.npy")
win = win / 10
# Imposta la directory
directory = "results/ST_arrays"

# Crea una nuova figura
plt.figure()

# Carica e traccia i dati per dist_5.npy
D_dist_5 = np.load(f"{directory}/sp_10_dist_5.npy")
D_dist_5 = D_dist_5[:-2] 
win_dist_5 = win[:-2] 
interp_func_dist_5 = interp1d(win_dist_5, D_dist_5, kind='linear')
win_new_dist_5 = np.linspace(min(win_dist_5), max(win_dist_5), 500)
D_interp_dist_5 = interp_func_dist_5(win_new_dist_5)
plt.scatter(win_dist_5, D_dist_5, label='$d_5$', s=10, color='red')
plt.plot(win_new_dist_5, D_interp_dist_5, color='red')

# Carica e traccia i dati per LENS_10.npy
D_LENS_10 = np.load(f"{directory}/sp_10_LENS_10.npy")
D_LENS_10 = D_LENS_10[:-2]+0.03
win_LENS = win[:-2]
interp_func_LENS_10 = interp1d(win_LENS, D_LENS_10, kind='linear')
win_new_LENS_10 = np.linspace(min(win_LENS), max(win_LENS), 500)
D_interp_LENS_10 = interp_func_LENS_10(win_new_LENS_10)
plt.scatter(win_LENS, D_LENS_10, label='LENS', s=10, color='blue')
plt.plot(win_new_LENS_10, D_interp_LENS_10, color='blue')

# Carica e traccia i dati per nn_10.npy
D_nn_10 = np.load(f"{directory}/sp_10_nn_10.npy")
D_nn_10 = D_nn_10[:-2]
win_NN = win[:-2]
interp_func_nn_10 = interp1d(win_NN, D_nn_10, kind='linear')
win_new_nn_10 = np.linspace(min(win_NN), max(win_NN), 500)
D_interp_nn_10 = interp_func_nn_10(win_new_nn_10)
plt.scatter(win_NN, D_nn_10, label='$NN_{10}$', s=10, color="brown")
plt.plot(win_new_nn_10, D_interp_nn_10, color="brown")

# Carica e traccia i dati per OTO.npy
D_OTO = np.load(f"{directory}/sp_10_OTO.npy")
D_OTO = D_OTO[:-2]
win_OTO = win[:-2]
interp_func_OTO = interp1d(win_OTO, D_OTO, kind='linear')
win_new_OTO = np.linspace(min(win_OTO), max(win_OTO), 500)
D_interp_OTO = interp_func_OTO(win_new_OTO)
plt.scatter(win_OTO, D_OTO, label='OTO', s=10, color = "black")
plt.plot(win_new_OTO, D_interp_OTO, color = "black")

# Carica e traccia i dati per SOAP_10_PC1.npy
D_SOAP_10_PC1 = np.load(f"{directory}/sp_10_SOAP_10_PC1.npy")
D_SOAP_10_PC1 = D_SOAP_10_PC1[:-2] -0.02
win_SOAP = win[:-2]
interp_func_SOAP_10_PC1 = interp1d(win_SOAP, D_SOAP_10_PC1, kind='linear')
win_new_SOAP_10_PC1 = np.linspace(min(win_SOAP), max(win_SOAP), 500)
D_interp_SOAP_10_PC1 = interp_func_SOAP_10_PC1(win_new_SOAP_10_PC1)
plt.scatter(win_SOAP, D_SOAP_10_PC1, label='PC1 (SOAP)', s=10, color="green")
plt.plot(win_new_SOAP_10_PC1, D_interp_SOAP_10_PC1, color="green")

# Carica e traccia i dati per vel.npy
D_vel = np.load(f"{directory}/sp_10_vel.npy")
D_vel = D_vel[:-2] +0.02
win_vel = win[:-2]
interp_func_vel = interp1d(win_vel, D_vel, kind='linear')
win_new_vel = np.linspace(min(win_vel), max(win_vel), 500)
D_interp_vel = interp_func_vel(win_new_vel)
plt.scatter(win_vel, D_vel, label='$v$', s=10, color="pink")
plt.plot(win_new_vel, D_interp_vel, color="pink")

# Aggiungi etichette agli assi
plt.xlabel('Time Resolution [ns]', fontweight='bold', fontsize=fontsize)
plt.xscale("log")
plt.ylabel('ENVs', fontweight='bold', fontsize=fontsize)
plt.tick_params(axis="x", labelsize=lsize)
plt.tick_params(axis="y", labelsize=lsize)

# Aggiungi un titolo al grafico
plt.title('Local noise reduction', fontsize=fontsize)
# Aggiungi una legenda
plt.legend(loc='lower left')
plt.ylim(0,4.1)
# Mostra il grafico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

directory = "results/ST_arrays"
# Example arrays (replace these with your actual data)
array1 = np.load(f"{directory}/dist_5.npy")
array2 = np.load(f"{directory}/sp_10_dist_5.npy")
array3 = np.load(f"{directory}/LENS_10.npy")
array4 = np.load(f"{directory}/sp_10_LENS_10.npy")
array5 = np.load(f"{directory}/nn_10.npy")
array6 = np.load(f"{directory}/sp_10_nn_10.npy")
array7 = np.load(f"{directory}/OTO.npy")
array8 = np.load(f"{directory}/sp_10_OTO.npy")
array9 = np.load(f"{directory}/SOAP_10_PC1.npy")
array10 = np.load(f"{directory}/sp_10_SOAP_10_PC1.npy")
array11 = np.load(f"{directory}/vel.npy")
array12 = np.load(f"{directory}/sp_10_vel.npy")

# Calculate maximum values of each array
max_values = [
    max(array3),
    max(array4),
    max(array1),
    max(array2),
    max(array7),
    max(array8),
    max(array9),
    max(array10),
    max(array5),
    max(array6),
    max(array11),
    max(array12),
]

# Define pairs of arrays
pairs = [
    (max_values[0], max_values[1]),
    (max_values[2], max_values[3]),
    (max_values[4], max_values[5]),
    (max_values[6], max_values[7]),
    (max_values[8], max_values[9]),
    (max_values[10], max_values[11])
]

# Labels for the pairs
pair_labels = ['LENS','$d_5$', 'OTO', 'SOAP', '$NN_{10}$','$v$']

# Prepare data for the histogram
values1 = [pair[0] for pair in pairs]
values2 = [pair[1] for pair in pairs]

# Create an array with indices for each group of bars
x = np.arange(len(pair_labels))

# Set the width of the bars
bar_width = 0.3

# Plot the histogram
fig, ax = plt.subplots(figsize=(6.2, 4.2))

# Plot bars for the first array in each pair
bars1 = ax.bar(x - bar_width/2, values1, bar_width, label='Raw data')

# Plot bars for the second array in each pair
bars2 = ax.bar(x + bar_width/2, values2, bar_width, label='Local noise reduction')

# Add labels, title, and legend
#ax.set_xlabel('Array Pairs')
ax.set_ylabel('ENVs', fontweight='bold', fontsize=fontsize)
ax.set_title('Max number of ENVs', fontsize=fontsize)
ax.set_xticks(x)
ax.set_xticklabels(pair_labels)
ax.legend()
plt.tick_params(axis="x", labelsize=lsize)
plt.tick_params(axis="y", labelsize=lsize)
plt.tight_layout()
plt.show()
```
